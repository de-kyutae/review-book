```python
import matplotlib.pyplot as plt
import numpy as np
from dezero import Function

class Sin(Function):
    def forward(self, x):
        """
        Forward of sin(x)
        
        returns: float
        """
        y = np.sin(x)
        return y
    
    def backward(self, gy):
        """
        derivation of sin(x) is equal to cos(x)
        gx : backward of sin(x), forward of sin(x) * derivation of sin(x)
        
        returns: float
        """
        x = self.inputs[0].data
        gx = gy * np.cos(x)
        return y
        
def sin(x):
    return Sin()(x)
```

```python
from dezero import Variable

'''when x= ğ¿/4, differentiate y=sin(x).'''
x = Variable(np.array(np.pi/4))
y = sin(x)
y.backward()

print(y.data)
print(x.grad.data) # xì˜ ë¯¸ë¶„ ê°’
```

```python
import math

def my_sin(x, threshold=0.0001):
    
    """ calculate talor series of sin(x) to 100000th derivation, unless (i)th derivation of sin(x) is smaller than threshold(0.0001)
    Args:
    threshold : absolut difference f(x) and f(x+ğ›¥)
    
    parameters :  
    c : (i)th of function [(-1)Ë†i/(2*i+1)!]
    t : (i)th derivation of sin(x), multiply (i)th of function c and (i)th of xË†(2*i+1)
    y : tayor series value of sin(x).
    
    """
    y = 0
    for i in range(100000):
        c = (-1)**i / math.factorial(2*i+1)
        t = c * x **(2*i+1)
        y = y+t
        if abs(t.data) < threshold:
            break
    return y
```

```python
x = Variable(np.array(np.pi/4))
y = my_sin(x)
y.backward()

print(y.data) #my_sin í•¨ìˆ˜ì— ì˜í•œ ê°’
print(x.grad.data) # xì˜ ë¯¸ë¶„ ê°’
```

```python
def rosenbrock(x0, x1):
    return 100*(x1 - x0**2) **2 + (1-x0)**2
```

```python
x0 = Variable(np.array(0.0))
x1 = Variable(np.array(2.0))

y = rosenbrock(x0, x1)
y.backward()
print(x0.grad, x1.grad)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits import mplot3d
import matplotlib.font_manager as mfm

plt.figure(figsize=(8,8))
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

x1 = np.linspace(-2, 2, 51)
x2 = np.linspace(-1,4, 51)
X1, X2 = np.meshgrid(x1,x2)


ax = plt.axes(projection='3d')
ax.plot_surface(X1, X2, rosenbrock(X1, X2), cmap='viridis', alpha=0.8)

ax.set_xlabel(r'$x_1$', fontsize=10)
ax.set_ylabel(r'$x_2$', fontsize=10)
ax.set_zlabel(r'$z$', fontsize=10)

plt.show()
```
